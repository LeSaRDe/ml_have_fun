```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Generate data with noise.
N = 500
w_0 = 1
w_1 = 2
X = np.random.uniform(low=-50, high=50, size=N)
E = np.random.normal(scale=20, size=N)
# U = np.random.uniform(low=-20, high=20, size=N)
U = np.random.normal(loc=-15, scale=10, size=N)
Y = w_0 + w_1 * X + E
```

```python
ax = plt.subplot()
ax.scatter(X, Y)
```

```python
D = np.stack((X, U), axis=-1)
D.shape
```

```python
D1_mesh, D2_mesh = np.meshgrid(D[:,0], D[:,1])
Y_mesh = w_0 + w_1 * D1_mesh + E
ax = plt.figure().add_subplot(projection='3d')
ax.plot_surface(D1_mesh, D2_mesh, Y_mesh, edgecolor='royalblue', lw=0.1, alpha=0.3)
```

```python
lin_reg = linear_model.LinearRegression().fit(D, Y)
score = lin_reg.score(D, Y)
print('Performance Score in Training =', score)
print('Coefficients =', lin_reg.coef_)
print('Intercept =', lin_reg.intercept_)
```
